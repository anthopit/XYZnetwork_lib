```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_excel('../data/Railway Data_JL.xlsx')
df
```

```python
G = nx.Graph()
G
```

```python
for index, row in df.iterrows():
    if not G.has_node(row["st_id"]):
        G.add_node(row["st_id"], lat=row["lat"], lon=row["lon"])
        
```
